```python
class Membrane(dolfin.SubDomain):
    def inside(self, x, on_boundary):
        r_check = numpy.sqrt(x[0]**2 + x[1]**2 + x[2]**2)
        r = 2.0
        eps = 0.5
        return r_check >= r-eps and r_check <= r+eps
    
class Ligand_Wall_Create(dolfin.SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and dolfin.near(x[2], 4.0)
    
class Ligand_Wall_Destroy(dolfin.SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and dolfin.near(x[2], -4.0)
    
class Extracellular_Space(dolfin.SubDomain):
    def inside(self, x, on_boundary):
        r_check = numpy.sqrt(x[0]**2 + x[1]**2 + x[2]**2)
        r = 2.0
        eps = 0.5
        return r_check > r+eps
    
class Extracellular_Space_Right(dolfin.SubDomain):
    def inside(self, x, on_boundary):
        r_check = numpy.sqrt(x[0]**2 + x[1]**2 + x[2]**2)
        r = 2.0
        eps = 0.5
        return r_check > r+eps and x[2]>=0
    
class Extracellular_Space_Left(dolfin.SubDomain):
    def inside(self, x, on_boundary):
        r_check = numpy.sqrt(x[0]**2 + x[1]**2 + x[2]**2)
        r = 2.0
        eps = 0.5
        return r_check > r+eps and x[2]<0
```

```python
class G_protein_cycle_1D(pyurdme.URDMEModel):

    def __init__(self,model_name="G_protein_cycle_1D"):
        pyurdme.URDMEModel.__init__(self,model_name)

        # Species
        # R RL G Ga Gbg Gd
        D_membrane = 0.0053   #diffusion on the membrane from PLOS paper
        D_bulk = 50           #not sure of bulk diffusion const
        D_L = 300
        L   = pyurdme.Species(name="L",  diffusion_constant=D_L)
        R   = pyurdme.Species(name="R",  diffusion_constant=D_membrane)
        RL  = pyurdme.Species(name="RL", diffusion_constant=D_membrane)
        G   = pyurdme.Species(name="G",  diffusion_constant=D_membrane)
        Ga  = pyurdme.Species(name="Ga", diffusion_constant=D_membrane)
        Gbg = pyurdme.Species(name="Gbg",diffusion_constant=D_membrane)
        Gd  = pyurdme.Species(name="Gd", diffusion_constant=D_membrane)
        self.add_species([L,R,RL,G,Ga,Gbg,Gd])
    
        #L = 4*3.14159
        #NUM_VOXEL = 200
        #MOLAR=6.02e-01*((L/NUM_VOXEL)**3)
        #self.mesh = pyurdme.URDMEMesh.generate_interval_mesh(nx=NUM_VOXEL, a=-2*3.14159, b=2*3.14159, periodic=True)
        #Define Geometry
        self.mesh = pyurdme.URDMEMesh(mesh = dolfin.Mesh('sphere_in_cube.xml'))
        
        cell_function = dolfin.CellFunction("size_t",self.mesh)
        cell_function.set_all(1)
        
        facet_function = dolfin.FacetFunction("size_t",self.mesh)
        facet_function.set_all(0)
        
        membrane = Membrane()
        membrane.mark(facet_function,2)
        
        extracellular_space = Extracellular_Space()
        extracellular_space.mark(facet_function,3)
        
#         ligand_wall_create = Ligand_Wall_Create()
#         ligand_wall_create.mark(facet_function,4)
        
#         ligand_wall_destroy = Ligand_Wall_Destroy()
#         ligand_wall_destroy.mark(facet_function,5)
        
#         extracellular_space_right = Extracellular_Space_Right()
#         extracellular_space_right.mark(facet_function,6)
        
#         extracellular_space_left = Extracellular_Space_Left()
#         extracellular_space_left.mark(facet_function,7)
        
        self.add_subdomain(cell_function)
        self.add_subdomain(facet_function)
          
        MOLAR = 1.493259e-04
        SA = 50.264
        #SA    = pyurdme.Parameter(name="SA" ,expression=50.264)
        V     = pyurdme.Parameter(name="V" ,expression=33.51)
        k_RL  = pyurdme.Parameter(name="k_RL" ,expression=2e-03/MOLAR)
        k_RLm = pyurdme.Parameter(name="k_RLm" ,expression=1e-02)
        k_Rs  = pyurdme.Parameter(name="k_Rs" ,expression=4.0/SA)
        k_Rd0 = pyurdme.Parameter(name="k_Rd0" ,expression=4e-04)
        k_Rd1 = pyurdme.Parameter(name="k_Rd1" ,expression=4e-04)
        k_G1  = pyurdme.Parameter(name="k_G1" ,expression=1.0*SA)
        k_Ga  = pyurdme.Parameter(name="k_Ga" ,expression=1e-05*SA)
        k_Gd  = pyurdme.Parameter(name="k_Gd" ,expression=0.1)
        k_Ls  = pyurdme.Parameter(name="k_Ls" ,expression=3.5e-02)
        k_Ld = pyurdme.Parameter(name="k_Ld" ,expression=4e-04)        
        self.add_parameter([V,k_RL,k_RLm,k_Rs,k_Rd0,k_Rd1,k_G1,k_Ga,k_Gd,k_Ls,k_Ld]) 

        # Add Data Function to model the mating pheromone gradient.
        #self.add_data_function(PheromoneGradient(L_mid=2, L_slope=1.0,L_max=10.0,L_min=0.0))

        # Reactions
        GR0 = pyurdme.Reaction(name="GR0", reactants={}, products={R:1}, massaction=True, rate=k_Rs, restrict_to=2)
        GR1 = pyurdme.Reaction(name="GR1", reactants={R:1}, products={}, massaction=True, rate=k_Rd0, restrict_to=2)
        GR2 = pyurdme.Reaction(name="GR2", reactants={R:1,L:1}, products={RL:1}, massaction=True, rate=k_RL, restrict_to=2)
        GR3 = pyurdme.Reaction(name="GR3", reactants={RL:1}, products={R:1,L:1}, massaction=True, rate=k_RLm, restrict_to=2)
        GR4 = pyurdme.Reaction(name="GR4", reactants={RL:1}, products={}, massaction=True, rate=k_Rd1, restrict_to=2)
        GR5 = pyurdme.Reaction(name="GR5", reactants={RL:1,G:1}, products={Ga:1, Gbg:1, RL:1}, massaction=True, rate=k_Ga, restrict_to=2)
        GR6 = pyurdme.Reaction(name="GR6", reactants={Ga:1}, products={Gd:1}, massaction=True, rate=k_Gd, restrict_to=2)
        GR7 = pyurdme.Reaction(name="GR7", reactants={Gd:1, Gbg:1}, products={G:1}, massaction=True, rate=k_G1, restrict_to=2)
        GR8 = pyurdme.Reaction(name="GR8", reactants={}, products={L:1}, propensity_function="(k_RL*R*L-k_RLm*RL)/L", restrict_to=3)
        GR9 = pyurdme.Reaction(name="GR9", reactants={L:1}, products={}, massaction=True, rate=k_Ld, restrict_to=5)
        self.add_reaction([GR0,GR1,GR2,GR3,GR4,GR5,GR6,GR7,GR8])#,GR9])
        
        self.restrict(species=R,subdomains=[2])
        self.restrict(species=RL,subdomains=[2])
        self.restrict(species=G,subdomains=[2])
        self.restrict(species=Ga,subdomains=[2])
        self.restrict(species=Gbg,subdomains=[2])
        self.restrict(species=Gd,subdomains=[2])
        self.restrict(species=L,subdomains=[2,3])
        
        
        # Distribute molecules randomly over the mesh according to their initial values
        self.set_initial_condition_scatter({R:10000},[2])
        self.set_initial_condition_scatter({G:10000},[2])
        self.set_initial_condition_scatter({L:5761},[3]) # 20nM micromolar through extracellular volume
#         self.set_initial_condition_scatter({L:200000},[6])
#         self.set_initial_condition_scatter({L:88050},[7])

        self.timespan(range(6000))
```

```python
def _connectivityGraph(mesh):
    mesh.init()
        
    # incidence relation between vertices and edges
    conn = mesh.topology()(0,mesh.topology().dim()-2)
    edg2vtx = mesh.topology()(mesh.topology().dim()-2,0)
        
    connectivity = []
    for vertex in range(mesh.num_vertices()):
        temp = []
        for edg in conn(vertex):
            for vtx in edg2vtx(edg):
                if vtx not in temp:
                    temp.append(vtx)

        temp.remove(vertex)
        connectivity.append(temp)

    return connectivity

neighbor_list = _connectivityGraph(model.mesh)
```
